# Data Cleaning and Preprocessing
Handle missing values, standardize responses, convert timestamp to datetime, and create age groups

```python
# Data Cleaning and Preprocessing

import pandas as pd

# Load the data
data = pd.read_csv('../data/cleaned/CleanedElectionSurvey_v3.csv',encoding='latin')

# Handle missing values
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Occupation'].fillna('Unknown', inplace=True)
data['Party_Belong'].fillna('Not Specified', inplace=True)

# Standardize responses
data['Reason_Not_Voted'].replace('Not Specified', 'No Response', inplace=True)
data['Reason_Voted_Candidate'].replace('Not Specified', 'No Response', inplace=True)
data['Reason_NPP_Lost'].replace('Not Specified', 'No Response', inplace=True)
data['Know_Projects'].replace('Not Specified', 'No Response', inplace=True)
data['Solved_Community_Problems'].replace('Not Specified', 'No Response', inplace=True)
data['Reason_NDC_Won'].replace('Not Specified', 'No Response', inplace=True)
data['NDC_Better_Than_NPP'].replace('Not Specified', 'No Response', inplace=True)
data['Best_Performing_MP'].replace('Not Specified', 'No Response', inplace=True)
data['Community_Name'].replace('Not Specified', 'No Response', inplace=True)
data['Biggest_Problems'].replace('Not Specified', 'No Response', inplace=True)
data['Addressed_Issues_Better'].replace('Not Specified', 'No Response', inplace=True)
data['Suggestions_2028'].replace('Not Specified', 'No Response', inplace=True)
data['Fair_Election_Results'].replace('Not Specified', 'No Response', inplace=True)

# Convert timestamp to datetime
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Create age groups
bins = [0, 18, 30, 45, 60, 100]
labels = ['<18', '18-30', '31-45', '46-60', '60+']
data['Age_Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Display the cleaned data
data.head()
```

```python
# Read and clean occupations data
def clean_occupations(data):
   # Convert to lowercase and strip whitespace
   data['Occupation'] = data['Occupation'].str.lower().str.strip()
   
   # Standardize common variations
   replacements = {
       'simstress': 'seamstress',
       'seanstress': 'seamstress', 
       'sramstress': 'seamstress',
       'hairdressee': 'hairdresser',
       'hairdtesser': 'hairdresser',
       'haidresser': 'hairdresser',
       'hairdersser': 'hairdresser',
       'fatmer': 'farmer',
       'okaada': 'okada rider',
       'okada riderridee': 'okada rider',
       'pensioneer': 'pensioner',
       'business man': 'businessman',
       'business woman': 'businesswoman',
       'unemplyed': 'unemployed'
   }
   data['Occupation'] = data['Occupation'].replace(replacements)
   
   # Define occupation categories
   categories = {
       'retail': ['trader', 'shop owner', 'shop assistant', 'store keeper', 
                 'provision store', 'boutique owner', 'food vendor'],
       'agriculture': ['farmer', 'poultry farmer', 'agriculturist', 'gardener'],
       'transportation': ['okada rider', 'driver'],
       'education': ['teacher', 'student'],
       'skilled_labor': ['carpenter', 'mason', 'electrician', 'plumber', 'welder',
                        'mechanic', 'factory worker'],
       'fashion_beauty': ['seamstress', 'tailor', 'hairdresser', 'barber'],
       'healthcare': ['nurse', 'pharmacist', 'physician assistant', 'midwife', 'herbalist'],
       'professional': ['banker', 'engineer', 'secretary', 'it specialist'],
       'unemployed': ['unemployed'],
       'unspecified': ['not specified']
   }
   
   # Map occupations to categories
   category_map = {}
   for category, occupations in categories.items():
       for occupation in occupations:
           category_map[occupation] = category
   
   data['Profession'] = data['Occupation'].map(category_map)
   data['Profession'] = data['Profession'].fillna('other')
   
   return data

# Group and summarize
def summarize_occupations(data):
   # Category totals
   category_totals = data.groupby('Profession')['Count'].sum().sort_values(ascending=False)
   
   # Top occupations per category
   top_by_category = data.groupby(['Profession', 'Occupation'])['Count'].sum()\
                         .sort_values(ascending=False)\
                         .groupby(level=0).head(3)
   
   return category_totals, top_by_category

# Example usage:
# df = pd.DataFrame({'Occupation': occupations, 'Count': counts})
df_cleaned = clean_occupations(df)
# cat_totals, top_occs = summarize_occupations(df_cleaned)

# print("Category Totals:")
# print(cat_totals)
# print("\nTop Occupations by Category:")
# print(top_occs)
```

```python
import pandas as pd
import re
import logging
from pathlib import Path

# Setup logging (if in a suitable environment)
# logging.basicConfig(
#     level=logging.INFO,
#     format='%(asctime)s - %(levelname)s - %(message)s'
# )

class ElectionDataCleaner:

    COLUMN_PATTERNS = {
        'Reason_Voted_Candidate': {
            r'(?i)liked\s*their\s*projects?\s*(or promises)?': 'Liked Projects/Promises',
            r'(?i)support\s*their\s*party': 'Supports Party',
            r'(?i)(for)?\s*change': 'For Change',
            r'(?i)active\s*in\s*the\s*community': 'Active in Community',
            r'(?i)(for)?\s*development': 'For Development',
            r'(?i)job\s*creation': 'Job Creation',
            r'(?i)peace': 'For Peace',
            r'(?i)low\s*standard\s*of\s*living|cost\s*of\s*living|economic\s*hardship|too\s*much\s*hardship|unemployment': 'Economic Hardship',
            r'(?i)gender\s*neutral\s*jobs': 'Gender Neutral Jobs',
            r'(?i)previous\s*performance|performed.*time.*office': 'Previous Performance',
            r'(?i)citizen': 'Citizen Duty',
            r'(?i)family\s*member|grandfather|closely\s*related|born\s*into\s*it': 'Family/Relationship',
            r'(?i)friend|convinced': 'Influenced by Others',
            r'(?i)better\s*government': 'Better Government',
            r'(?i)currency\s*exchange|loans|bet\s*taxes|remove\s*betting\s*tax': 'Specific Issues',
            r'(?i)forced|father': 'Forced to Vote',
            r'(?i)award': 'Award',
            r'(?i)no\s*response|^$': 'No Response',  # Match empty strings as well
            r'(?i).*': 'Other' # Catch-all to ensure all responses are categorized
        },
        'Party_Belong': {
            r'(?i)^NDC\s*$': 'NDC',
            r'(?i)^NPP\s*$': 'NPP',
            r'(?i)^CPP\s*$': 'CPP',
            r'(?i)^LPG\s*$': 'LPG',
            r'(?i)new\s*force': 'New Force',
            r'(?i)none|unknown|cant\s*tell': 'Not Specified'
            r'(?i).*': 'Other' # Catch-all to ensure all responses are categorized
            
    }

    def standardize_column(self, series):
        """
        Standardizes a Pandas Series based on defined regex patterns.

        Args:
            series: The Pandas Series to standardize.

        Returns:
            The standardized Pandas Series.
        """

        def replace_with_first_match(text):
            for pattern, replacement in self.COLUMN_PATTERNS['Reason_Voted_Candidate'].items():
                if re.search(pattern, text, re.IGNORECASE):
                    return replacement
            return 'Other'  # Should not reach here due to catch-all, but added for safety

        return series.apply(replace_with_first_match)


    def clean_and_group(self, input_series):
        """
        Cleans, standardizes, and groups responses in a Pandas Series.

        Args:
            input_series (pd.Series): The input series containing the responses.

        Returns:
            pd.Series: A cleaned, standardized, and grouped series.
        """

        cleaned_series = self.standardize_column(input_series)
        grouped_counts = cleaned_series.value_counts()

        return grouped_counts

# Example usage (replace with your actual series)
cleaner = ElectionDataCleaner()
# df['Reason_Voted_Candidate_Standardized'] = cleaner.standardize_column(df['Reason_Voted_Candidate'])
result = cleaner.clean_and_group(df['Reason_Voted_Candidate'])
print(result)
# Create a new standardized column by applying the cleaner
```
